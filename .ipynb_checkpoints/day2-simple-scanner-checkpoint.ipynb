## Day 2 - Simple Port Scanner using python-nmap
### Safe & legal target: scanme.nmap.org (permission granted by nmap.org)

```python
import nmap
```

```python
def scan_target(target):
    print(f"[+] Starting scan on {target}")
    print("[+] This may take 20-60 seconds...\n")
    
    scanner = nmap.PortScanner()
    # -F = Fast scan (top 100 ports), -sV = version detection
    result = scanner.scan(target, arguments='-F -sV')
    
    print("="*50)
    print(f"Host: {target} ({scanner[target].hostname()})")
    print(f"State: {scanner[target].state()}")
    print("="*50)
    
    for proto in scanner[target].all_protocols():
        ports = scanner[target][proto].keys()
        for port in sorted(ports):
            state = scanner[target][proto][port]['state']
            service = scanner[target][proto][port]['name']
            version = scanner[target][proto][port].get('product', '')
            version_info = f" ({version})" if version else ""
            print(f"Port {port}/tcp   {state:8}   {service}{version_info}")

if __name__ == "__main__":
    # This target is explicitly allowed for testing
    scan_target("scanme.nmap.org")
```
